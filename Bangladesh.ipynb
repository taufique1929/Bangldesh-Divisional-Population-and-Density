```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bd_cities = pd.read_csv('BD_cities.csv')
```

```python
lat, lon = bd_cities['Latitude'], bd_cities['Longitude']
pop, area = bd_cities['Population'], bd_cities['Area']
den = bd_cities['Density']
```

```python
plt.figure(figsize=(20,20))
plt.scatter(lon, lat, label=None, c=np.log10(pop), cmap='viridis', s=area, linewidth=0, alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='log$_{10}$(Population)')
plt.clim(7,8)
plt.xlim(87,93)
plt.ylim(21,27)

for area in [10000, 20000, 30000]:
    plt.scatter([], [], c='k', alpha=0.3, s=area,
                       label=str(area) + ' km$^2$')
    plt.legend(scatterpoints=1, frameon=False,
                  labelspacing=5, title='Division Area')

plt.title('Bangladesh Divisions: Area and Populations')
plt.savefig('BD_divisions_population_area')
```

```python
plt.figure(figsize=(20,20))
plt.scatter(lon, lat, label=None, c=den, cmap='viridis', s=area, linewidth=0, alpha=1.0)
plt.grid()
plt.axis(aspect='equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Density')
#plt.clim(7,8)
plt.xlim(87,93)
plt.ylim(21,27)

for area in [10000, 20000, 30000]:
    plt.scatter([], [], c='r', alpha=0.3, s=area,
                       label=str(area) + ' km$^2$')
    plt.legend(scatterpoints=1, frameon=False,
                  labelspacing=5, title='Division Area')

plt.title('Bangladesh Divisions: Area and Populations')
plt.savefig('BD_divisions_density_area')
```
